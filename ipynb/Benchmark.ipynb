```python

plt.style.use('default')

def plot_mean(ax, data, color='b'):
    ax.axhline(y=np.mean(data), color=color, linestyle='--')

# Define a function to plot and save each metric
def plot_and_save_julia(metric, ylabel, filename):
    fig, ax = plt.subplots(figsize=(5, 4))
    ax.plot(SMALL_GPU_JULIA["epoch"], SMALL_GPU_JULIA[metric], label='SMALL', color='c')
    ax.plot(MEDIUM_GPU_JULIA["epoch"], MEDIUM_GPU_JULIA[metric], label='MEDIUM', color='m')
    ax.plot(BIG_GPU_JULIA["epoch"], BIG_GPU_JULIA[metric], label='BIG', color='y')


    plot_mean(ax, SMALL_GPU_JULIA[metric], 'c')
    plot_mean(ax, MEDIUM_GPU_JULIA[metric], 'm')
    plot_mean(ax, BIG_GPU_JULIA[metric], 'y')

    ax.plot(0, SMALL_GPU_JULIA[metric][0], color='r', linestyle='--', label='Mean Value')
    


    ax.set(xlabel='Epoch', ylabel=ylabel)
    ax.legend()
    plt.tight_layout()
    plt.savefig(f"../image/JULIA_{filename}")

    plt.close()


def plot_and_save_python(metric, ylabel, filename):
    fig, ax = plt.subplots(figsize=(5, 4))
    ax.plot(SMALL_GPU_PYTHON["epoch"], SMALL_GPU_PYTHON[metric], label='SMALL', color='c')
    ax.plot(MEDIUM_GPU_PYTHON["epoch"], MEDIUM_GPU_PYTHON[metric], label='MEDIUM', color='m')
    ax.plot(BIG_GPU_PYTHON["epoch"], BIG_GPU_PYTHON[metric], label='BIG', color='y')


    plot_mean(ax, SMALL_GPU_PYTHON[metric], 'c')
    plot_mean(ax, MEDIUM_GPU_PYTHON[metric], 'm')
    plot_mean(ax, BIG_GPU_PYTHON[metric], 'y')

    ax.plot(0, SMALL_GPU_PYTHON[metric][0], color='r', linestyle='--', label='Mean Value')



    ax.set(xlabel='Epoch', ylabel=ylabel)
    ax.legend()
    plt.tight_layout()
    plt.savefig(f"../image/PYTHON_{filename}")
    plt.close()

plot_and_save_julia("epoch_memory_usage", "Memory Usage", "memory_usage.png")
plot_and_save_julia("epoch_time", "Epoch Time", "epoch_time.png")
plot_and_save_julia("throughput", "Throughput", "throughput.png")
plot_and_save_julia("accuracy", "Accuracy", "accuracy.png")
plot_and_save_julia("loss", "Loss", "loss.png")

plot_and_save_python("epoch_memory_usage", "Memory Usage", "memory_usage.png")
plot_and_save_python("epoch_time", "Epoch Time", "epoch_time.png")
plot_and_save_python("throughput", "Throughput", "throughput.png")
plot_and_save_python("accuracy", "Accuracy", "accuracy.png")
plot_and_save_python("loss", "Loss", "loss.png")
```

```python
def plot_mean(ax, data, color='b'):
    ax.axhline(y=np.mean(data), color=color, linestyle='--', label='Mean Value')

# Define a function to plot and save each metric
def plot_and_save_julia(metric, ylabel, filename):
    fig, ax = plt.subplots(figsize=(5, 4))
    ax.plot(SMALL_GPU_JULIA["epoch"], SMALL_GPU_JULIA[metric], label='SMALL', color='c')
    plot_mean(ax, SMALL_GPU_JULIA[metric], 'c')

    ax.set(xlabel='Epoch', ylabel=ylabel)
    ax.legend()
    plt.tight_layout()
    plt.savefig(f"../image/JULIA_SMALL_{filename}")

    ax.plot(MEDIUM_GPU_JULIA["epoch"], MEDIUM_GPU_JULIA[metric], label='MEDIUM', color='m')
    plot_mean(ax, MEDIUM_GPU_JULIA[metric], 'm')

    ax.set(xlabel='Epoch', ylabel=ylabel)
    ax.legend()
    plt.tight_layout()
    plt.savefig(f"../image/JULIA_MEDIUM_{filename}")

    ax.plot(BIG_GPU_JULIA["epoch"], BIG_GPU_JULIA[metric], label='BIG', color='y')
    plot_mean(ax, BIG_GPU_JULIA[metric], 'y')


    ax.set(xlabel='Epoch', ylabel=ylabel)
    ax.legend()
    plt.tight_layout()
    plt.savefig(f"../image/JULIA_BIG_{filename}")

def plot_and_save_python(metric, ylabel, filename):
    fig, ax = plt.subplots(figsize=(5, 4))
    ax.plot(SMALL_GPU_PYTHON["epoch"], SMALL_GPU_PYTHON[metric], label='SMALL', color='c')
    plot_mean(ax, SMALL_GPU_PYTHON[metric], 'c')

    ax.set(xlabel='Epoch', ylabel=ylabel)
    ax.legend()
    plt.tight_layout()
    plt.savefig(f"../image/PYTHON_SMALL_{filename}")

    ax.plot(MEDIUM_GPU_PYTHON["epoch"], MEDIUM_GPU_PYTHON[metric], label='MEDIUM', color='m')
    plot_mean(ax, MEDIUM_GPU_PYTHON[metric], 'm')

    ax.set(xlabel='Epoch', ylabel=ylabel)
    ax.legend()
    plt.tight_layout()
    plt.savefig(f"../image/PYTHON_MEDIUM_{filename}")

    ax.plot(BIG_GPU_PYTHON["epoch"], BIG_GPU_PYTHON[metric], label='BIG', color='y')
    plot_mean(ax, BIG_GPU_PYTHON[metric], 'y')

    ax.set(xlabel='Epoch', ylabel=ylabel)
    ax.legend()
    plt.tight_layout()
    plt.savefig(f"../image/PYTHON_BIG_{filename}")



plot_and_save_julia("epoch_memory_usage", "Memory Usage", "memory_usage.png")
plot_and_save_python("epoch_memory_usage", "Memory Usage", "memory_usage.png")
```
